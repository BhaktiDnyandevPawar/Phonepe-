### Dataset Loading

````python
path = '/content/pulse/data/aggregated/transaction/country/india/state/'

agg_trans_list = os.listdir(path)

data = {
    "State": [],
    "Year": [],
    "Quarter": [],
    "Transaction_Type": [],
    "Transaction_Count": [],
    "Transaction_Amount": []
}

for state in agg_trans_list:
    state_path = path + state + "/"
    year_list = os.listdir(state_path)

    for year in year_list:
        year_path = state_path + year + "/"
        file_list = os.listdir(year_path)

        for file in file_list:
            file_path = year_path + file

            with open(file_path, 'r') as f:
                content = json.load(f)

                if content['data'] is not None:
                    for i in content['data']['transactionData']:
                        data["State"].append(state)
                        data["Year"].append(year)
                        data["Quarter"].append(int(file.strip('.json')))
                        data["Transaction_Type"].append(i['name'])
                        data["Transaction_Count"].append(i['paymentInstruments'][0]['count'])
                        data["Transaction_Amount"].append(i['paymentInstruments'][0]['amount'])
````

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Convert Year to integer
df['Year'] = df['Year'].astype(int)

# Convert State names to proper format (capitalize)
df['State'] = df['State'].str.replace('-', ' ').str.title()

# Check and remove duplicates (already 0 but safe)
df = df.drop_duplicates()

# Handle missing values (if any)
df.fillna(0, inplace=True)

# Sort data for better analysis
df = df.sort_values(by=['Year', 'Quarter'])

# Reset index
df.reset_index(drop=True, inplace=True)

# Final check
df.head()
````

### What all manipulations have you done and insights you found?

In the data wrangling process, several transformations were performed to make the dataset suitable for analysis. The “Year” column was converted from object type to integer to enable proper time-based analysis. The “State” column was cleaned by replacing hyphens with spaces and converting text into a readable format.

Duplicate values were checked and removed to maintain data accuracy. Missing values were handled appropriately to avoid inconsistencies during analysis. The dataset was then sorted based on year and quarter to maintain chronological order, and the index was reset for better structure.

###Create Database Connection

````python
import sqlite3

conn = sqlite3.connect("phonepe.db")
cursor = conn.cursor()
````

###Store DataFrame into SQL

````python
df.to_sql("transactions", conn, if_exists="replace", index=False)
````

###Check Data

````python
pd.read_sql("SELECT * FROM transactions LIMIT 5", conn)
````

###SQL Queries

````python
#Query 1: Top States
query1 = """
SELECT State, SUM(Transaction_Amount) AS Total_Amount
FROM transactions
GROUP BY State
ORDER BY Total_Amount DESC
LIMIT 10;
"""

pd.read_sql(query1, conn)
````

````python
#Query 2: Transaction Type Analysis
query2 = """
SELECT Transaction_Type, SUM(Transaction_Count) AS Total_Count
FROM transactions
GROUP BY Transaction_Type;
"""

pd.read_sql(query2, conn)
````

````python
#Query 3: Year-wise Growth
query3 = """
SELECT Year, SUM(Transaction_Amount) AS Total_Amount
FROM transactions
GROUP BY Year
ORDER BY Year;
"""

pd.read_sql(query3, conn)
````

SQL was used to store the processed dataset into a structured database using SQLite in Google Colab. The dataset was converted into a table and various SQL queries such as SELECT, GROUP BY, and ORDER BY were executed to analyze transaction trends, identify top-performing states, and understand transaction categories. This improved data handling efficiency and enabled quick retrieval of insights
